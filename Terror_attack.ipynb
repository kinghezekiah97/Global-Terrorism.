Global Terrorism Attack Analysis

```python
#IMPORT LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
terror_data = pd.read_csv("Desktop/Data Analysis class/Python-class/globalterrorismdb_0718dist.csv", encoding='ISO-8859-1', low_memory=False)
```

```python
#Data overview
terror_data.head
```

```python
# top ten countries affected by terrorist attacks
top_countries = terror_data['country_txt'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, color='blue')
plt.title('Top 10 Countries by Number of Terrorist Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.show()
```

Iran, Pakistan and Afghanistan experience the highest number of terrorist attack

```python
# most common terrorist attack type
attack_types = terror_data['attacktype1_txt'].value_counts().head(10) #top 10 most attack type
plt.figure(figsize=(10, 6)) 
sns.barplot(x=attack_types.values, y=attack_types.index, color='purple')
plt.title('Most Common Terrorist Attack Types')
plt.xlabel('Number of Attacks')
plt.ylabel('Attack Type')
plt.show()
```

Bombing/explosion, Armed assault and assination are common attack type used by terrorist

```python
#trend of terrorist attack between 1970-2017
attacks_per_year = terror_data['iyear'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x=attacks_per_year.index, y=attacks_per_year.values, marker='o')
plt.title('Terrorist Attacks Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

There is increase in terrorist attack from 2000 to 2010, but there is decrease in terrorist attack between  2016-2018

```python
#top 10 most active terrorist group
top_groups = terror_data[terror_data['gname'] != 'Unknown']['gname'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_groups.values, y=top_groups.index, palette='Reds')
plt.title('Top 10 Most Active Terrorist Organizations')
plt.xlabel('Number of Attacks')
plt.ylabel('Group Name')
plt.show()
```

```python
Taliban and islamic state of iraq and the levant(ISIL) are the most active group
```

```python
# number of attack month by month(to show seasonality

monthly_attacks = terror_data['imonth'].value_counts().sort_index()
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(10, 6))
sns.barplot(x=months, y=monthly_attacks[1:13], color='red')  
plt.title('Terrorist Attacks by Month')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.show()
```

Most attack occur in narch,may,july which shows a seasonality in the terrorist attack

```python
#most active terrorist group in nigeria
nigeria_terror = terror_data[terror_data['country_txt'] == 'Nigeria'] #select only nigeria fom country_txt
top_group_nigeria = nigeria_terror['gname'].value_counts().drop('Unknown').head(5) # select gname when country is nigeria, and show top 10

print("Most Active Group in Nigeria:", top_group_nigeria)
```

Boko Haram dominates most terrorist attack in Nigeria

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_group_nigeria.values, y=top_group_nigeria.index, color='violet')
plt.title('Top 5 Most Active Terrorist in Nigeria')
plt.xlabel('Number of Attacks')
plt.ylabel('Group Name')
plt.show()
```

```python
# top 5 target in nigeria using df from preceding code
top_target_nigeria = nigeria_terror['targtype1_txt'].value_counts().head(5) #  count number of target type from targtype1_txt from nigeria terror and show top 5
print("Most Targeted Type in Nigeria:", top_target_nigeria)
```

```python
# barchat for top 5 target in nigeria using df from preceding code 
plt.figure(figsize=(10, 6))
sns.barplot(x=top_target_nigeria.values, y=top_target_nigeria.index, color='orange') #barplot from seaborn
plt.title('Top 5 Target in Nigeria')
plt.xlabel('Number of Attacks')
plt.ylabel('Targets')
plt.show()
```
